```python
df2 = df.loc[df['device'].str.endswith("Person0/elements/beta_absolute")]
df2 = df2.reset_index()
del df2['index']
df2.head()
df2['second'] = df2['second'].astype(float)
print(df2.dtypes)
df2['Value'] = (df2['first'] + df2['second'] + df2['third'] + df2['fourth']) /4
dfout = df2.drop(columns = ['first', 'second', 'third', 'fourth'])
dfout['device'] = "beta_absolute"
```
